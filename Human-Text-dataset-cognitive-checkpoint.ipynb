```python
import pandas as pd

from datasets import load_dataset
```

```python
ds= load_dataset("CollectiveCognition/chats-data-2023-09-27")
print(ds['train'].features)
```

```python
flattened =[]

for row in ds['train']:
    for msg in row['conversations']:
        flattened.append({
            'speaker': msg['from'],
            'text': msg['value'],
            'model_name': row['model_name'],
            'title': row['title']
        })


    df= pd.DataFrame(flattened)
    print(df.head())
```

```python
df['text'] = df['text'].astype(str)
df['text_length'] = df['text'].apply(lambda x: len(x.split()))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='speaker', order=df['speaker'].value_counts().index, palette='Set2')
plt.title("Number of Messages by Speaker")
plt.xlabel("Speaker")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='text_length', hue='speaker', bins=40, kde=True, element='step')
plt.title("Message Length Distribution by Speaker")
plt.xlabel("Words per Message")
plt.ylabel("Frequency")
plt.xlim(0, 100)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.countplot(data=df, y='model_name', order=df['model_name'].value_counts().index, palette='coolwarm')
plt.title("Number of Messages per Model")
plt.xlabel("Count")
plt.ylabel("Model")
plt.tight_layout()
plt.show()
```

```python
human_df = df[df['speaker'].str.lower() == 'human'].reset_index(drop=True)
```
